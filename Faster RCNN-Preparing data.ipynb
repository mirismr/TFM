```
import numpy as np
import time
import sys
import os
import random
from skimage import io
import pandas as pd
from shutil import copyfile
import cv2
from google.colab import drive
drive.mount('/content/drive')


images_boxable_fname = '/content/drive/My Drive/train-images-boxable.csv'
annotations_bbox_fname = '/content/drive/My Drive/train-annotations-bbox.csv'
class_descriptions_fname = '/content/drive/My Drive/class-descriptions-boxable.csv'


images_boxable = pd.read_csv(images_boxable_fname)
annotations_bbox = pd.read_csv(annotations_bbox_fname)
class_descriptions = pd.read_csv(class_descriptions_fname)


horse_pd = class_descriptions[(class_descriptions[['class']]=='Horse').values.tolist()]
car_pd = class_descriptions[(class_descriptions[['class']]=='Car').values.tolist()]
dog_pd = class_descriptions[(class_descriptions[['class']]=='Dog').values.tolist()]
person_pd = class_descriptions[(class_descriptions[['class']]=='Person').values.tolist()]


label_name_horse = horse_pd['name'].values[0]
label_name_car = car_pd['name'].values[0]
label_name_dog = dog_pd['name'].values[0]
label_name_person = person_pd['name'].values[0]


horse_bbox = annotations_bbox[annotations_bbox['LabelName']==label_name_horse]
car_bbox = annotations_bbox[annotations_bbox['LabelName']==label_name_car]
dog_bbox = annotations_bbox[annotations_bbox['LabelName']==label_name_dog]
person_bbox = annotations_bbox[annotations_bbox['LabelName']==label_name_person]




horse_img_id = np.unique(horse_bbox['ImageID'])
car_img_id = np.unique(car_bbox['ImageID'])
dog_img_id = np.unique(dog_bbox['ImageID'])
person_img_id = np.unique(person_bbox['ImageID'])


n = 1000
index_horse = np.random.randint(0, len(horse_img_id), n)
index_car = np.random.randint(0, len(car_img_id), n)
index_dog = np.random.randint(0, len(dog_img_id), n)
index_person = np.random.randint(0, len(person_img_id), n)


horse_data = horse_img_id[index_horse]
car_data = car_img_id[index_car]
dog_data = dog_img_id[index_dog]
person_data = person_img_id[index_person]


horse_img_url = [images_boxable[images_boxable['image_name']==name+'.jpg'] for name in horse_data]
car_img_url = [images_boxable[images_boxable['image_name']==name+'.jpg'] for name in car_data]
dog_img_url = [images_boxable[images_boxable['image_name']==name+'.jpg'] for name in dog_data]
person_img_url = [images_boxable[images_boxable['image_name']==name+'.jpg'] for name in person_data]


urls = [horse_img_url, car_img_url, dog_img_url, person_img_url]


print(len(urls[0]))
print(urls[0][0])
```

```
classes = ['Horse', 'Car', 'Dog', 'Person']
for i in range(len(classes)):
    # Create the directory
    dir = '/content/drive/My Drive/'+classes[i]
    if (not os.path.isdir(dir)):
      os.mkdir(dir)
    saved_dir = dir
    for url in urls[i]:
        url_aux = url['image_url'].values[0]
        try:
          img = io.imread(url_aux)
        except Exception as e:
          print(url_aux)
        
        saved_path = os.path.join(saved_dir, url_aux[-20:])
        io.imsave(saved_path, img)
```

```
dir = '/content/drive/My Drive/'
classes = ['Horse', 'Car', 'Dog', 'Person']

train_path = dir+'train/'
test_path = dir +'test/'
if (not os.path.isdir(train_path)): 
  os.mkdir(train_path)
if (not os.path.isdir(test_path)):
  os.mkdir(test_path)

for i in range(len(classes)):
    class_path = dir+classes[i]
    all_imgs = os.listdir(class_path)
    all_imgs = [f for f in all_imgs if not f.startswith('.')]
    random.seed(1)
    random.shuffle(all_imgs)
    
    train_imgs = all_imgs[:800]
    test_imgs = all_imgs[800:1001]
    
    # Copy each classes' images to train directory
    for j in range(len(train_imgs)):
        original_path = class_path+"/"+train_imgs[j]
        new_path = train_path+train_imgs[j]
        copyfile(original_path, new_path)
    
    # Copy each classes' images to test directory
    for j in range(len(test_imgs)):
        original_path = class_path+"/"+test_imgs[j]
        new_path = test_path+test_imgs[j]
        copyfile(original_path, new_path)
```

```
dir = '/content/drive/My Drive/'
print('number of train images: ', len(os.listdir(train_path)))
print('number of test images: ', len(os.listdir(test_path)))
```

```
label_names = [label_name_horse, label_name_car, label_name_dog, label_name_person]

train_df = pd.DataFrame(columns=['FileName', 'XMin', 'XMax', 'YMin', 'YMax', 'ClassName'])

# Find boxes in each image and put them in a dataframe
train_imgs = os.listdir(train_path)
train_imgs = [name for name in train_imgs if not name.startswith('.')]

for i in range(len(train_imgs)):
    sys.stdout.write('Parse train_imgs ' + str(i) + '; Number of boxes: ' + str(len(train_df)) + '\r')
    sys.stdout.flush()
    img_name = train_imgs[i]
    img_id = img_name[0:16]
    tmp_df = annotations_bbox[annotations_bbox['ImageID']==img_id]
    for index, row in tmp_df.iterrows():
        labelName = row['LabelName']
        for i in range(len(label_names)):
            if labelName == label_names[i]:
                train_df = train_df.append({'FileName': img_name, 
                                            'XMin': row['XMin'], 
                                            'XMax': row['XMax'], 
                                            'YMin': row['YMin'], 
                                            'YMax': row['YMax'], 
                                            'ClassName': classes[i]}, 
                                           ignore_index=True)
```

```
test_df = pd.DataFrame(columns=['FileName', 'XMin', 'XMax', 'YMin', 'YMax', 'ClassName'])

# Find boxes in each image and put them in a dataframe
test_imgs = os.listdir(test_path)
test_imgs = [name for name in test_imgs if not name.startswith('.')]

for i in range(len(test_imgs)):
    sys.stdout.write('Parse test_imgs ' + str(i) + '; Number of boxes: ' + str(len(test_df)) + '\r')
    sys.stdout.flush()
    img_name = test_imgs[i]
    img_id = img_name[0:16]
    tmp_df = annotations_bbox[annotations_bbox['ImageID']==img_id]
    for index, row in tmp_df.iterrows():
        labelName = row['LabelName']
        for i in range(len(label_names)):
            if labelName == label_names[i]:
                test_df = test_df.append({'FileName': img_name, 
                                            'XMin': row['XMin'], 
                                            'XMax': row['XMax'], 
                                            'YMin': row['YMin'], 
                                            'YMax': row['YMax'], 
                                            'ClassName': classes[i]}, 
                                           ignore_index=True)
```

```

train_df.to_csv(dir+'train.csv')
test_df.to_csv(dir+'test.csv')
```
